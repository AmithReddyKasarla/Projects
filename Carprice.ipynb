```python
df=df[['year', 'selling_price', 'km_driven', 'fuel', 'seller_type',
       'transmission', 'owner']]
```

```python
df['current_year']=2022
```

```python
!pip install sklearn
```

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
ex = ExtraTreesRegressor()
```

```python
ex.fit(X,y)
```

```python
print(ex.feature_importances_)
```

```python
fea_imp = pd.Series(ex.feature_importances_, index=X.columns)
fea_imp.nlargest(3).plot()
plt.show()
```

```python
fea_imp = pd.Series(ex.feature_importances_, index=X.columns)
fea_imp.nlargest(3).plot(kind = 'bar')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.30)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=200,random_state = 25, max_depth = 5, oob_score=True)
```

```python
rf.fit(X_train,y_train)
```

```python
pr = rf.predict(X_test)
```

```python
pr
```

```python
sns.distplot(y_test-pr)
```

```python
plt.scatter(y_test,pr)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits = 5, shuffle = True, random_state = 5)
```

```python
kf
```

```python
score = cross_val_score(rf,X_train,y_train, cv = kf)
```
